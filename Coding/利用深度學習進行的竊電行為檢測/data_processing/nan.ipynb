遺失值處理

```python
#broadcasting
# 將第三行缺失值填充為第四行的值(如果第四行的缺失值為缺失值，則第三行依舊為缺失值，在填充0)
df_test.iloc[:, 1].fillna(df_test.iloc[:, 2], inplace=True)
df_test.iloc[:, 1].fillna(0, inplace=True)

# 將最後一行缺失值填充為倒數第二行的值
df_test.iloc[:, -1].fillna(df_test.iloc[:, -2], inplace=True)
df_test.iloc[:, -1].fillna(0, inplace=True)

for i in range(3, len(df_test.columns) - 1):
    df_test.iloc[:, i].fillna((df_test.iloc[:, i-1] + df_test.iloc[:, i+1]) / 2, inplace=True)
    df_test.iloc[:, i].fillna(0, inplace=True)

output_filename = 'nonan_test.csv'
df_test.to_csv(output_filename, index=False)


# for j in range(len(df_test['CONS_NO'])):
#     if pd.isnull(df_test.iloc[j,1]) :
#         if pd.isnull(df_test.iloc[j,2]):
#                 df_test.iloc[j,1] = 0
#                 # print(df.iloc[j,1])
#         else:
#             df_test.iloc[j,1] = df_test.iloc[j,2]
#     if pd.isnull(df_test.iloc[j,-1]) :
#         if pd.isnull(df_test.iloc[j,-2]):
#                 df_test.iloc[j,-1] = 0
#                 # print(df.iloc[j,1])
#     for i in range(2, len(df_test.columns)-1):
#         if pd.isnull(df_test.iloc[j,i]) :
#             if pd.isnull(df_test.iloc[j,i-1]) | pd.isnull(df_test.iloc[j,i+1]):
#                 df_test.iloc[j,i] = 0
#                 # print(df.iloc[j,i])
#             else:
#                 df_test.iloc[j,i] = df_test.iloc[j,i-1] + df_test.iloc[j,i+1] / 2
#                 # print(df.iloc[j,i])
#         else:
#             df_test.iloc[j,i] = df_test.iloc[j,i]
#             # print(df.iloc[j,i])
```
