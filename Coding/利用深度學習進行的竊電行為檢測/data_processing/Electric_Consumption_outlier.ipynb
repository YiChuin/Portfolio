```python
import pandas as pd

# 从 CSV 文件中读取数据并加载到 DataFrame 中
df = pd.read_csv('electricitytheft_train.csv')

# 显示 DataFrame 的前几行
df.head(5)
```

```python
# 从 DataFrame 中提取日期列
df_date = df.iloc[:,2:]
df_other = df.iloc[:,0:2]

# 将第一列作为索引列
df_date.set_index(df_date.columns[0], inplace=True)

# 解析日期并转置数据
df_date.columns = pd.to_datetime(df_date.columns)
df_date = df_date.T

# 按照日期对数据进行排序
df_date = df_date.sort_index()
# 再次转置数据并重置索引
df_date = df_date.T.reset_index()

df = pd.concat([df_other,df_date], axis=1)
```

計算各行標準差平均值等等

```python
# 從第三行開始計算，忽略前兩行
df_to_calculate = df.iloc[:,2:]


# 計算每一列的變異數
row_variances = df_to_calculate.var(axis=1)

# 計算每一列的平均值和標準差
row_means = df_to_calculate.mean(axis=1)
row_std = df_to_calculate.std(axis=1)
```

對前20筆做調整測試

```python
import numpy as np
threshold = row_means + 2 * row_std

for i in range(20):
    row = df_to_calculate.iloc[i]
    #row_numeric = pd.to_numeric(row, errors='coerce')  # 將字符串值轉換為數字，無法轉換的部分設為NaN
    df_to_calculate.iloc[i + 2, :] = np.where(row > threshold[i], row_means[i] + 2 * row_std[i], row)



```

```python
import numpy as np

# 初始化計數器
count_values_exceeding_threshold = 0

# 計算每一行的閾值，並統計超過閾值的值的數量
for i in range(df_to_calculate.shape[0]):
    row = df_to_calculate.iloc[i]
    threshold = row_means[i] + 2 * row_std[i]
    count_values_exceeding_threshold += np.sum(row > threshold)

print("Total values exceeding mean plus two standard deviations:", count_values_exceeding_threshold)
```
